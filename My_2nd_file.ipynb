```python
pd.crosstab(tips['smoker'], tips['sex'])
```

```python
pd.crosstab(tips['smoker'], tips['sex']).plot(kind = 'bar')
```

```python
#strip plot
import matplotlib.pyplot as plt
sns.stripplot(y = 'tip', data = tips, jitter = True)
plt.ylable('tip ($)')
plt.show()
```

```python
#strip plot
sns.stripplot(x = 'sex', y = 'tip', data = tips, jitter = True)
plt.title('Strip plot')
plt.xlable('Gender')
plt.show()
```

```python
#Grouping with stripplot()
sns.stripplot(x='day', y='tip', data=tips)
plt.ylable('tip ($)')
plt.show()
```

```python
sns.stripplot(x='sex', y='tip', hue='day', data=tips)
plt.ylable('tip ($)')
plt.show()
```

```python
sns.stripplot(x='day', y='tip', hue='sex', data=tips, size = 5)
plt.ylable('tip ($)')
plt.show()
```

```python
#Swarm plot
sns.swarmplot(x='day', y='tip', data=tips, size = 5)
plt.ylable('tip ($)')
plt.show()
```

```python
tips['sex'].unique()
```

```python
#Swarm plot
sns.swarmplot(x='day', y='tip', data=tips, hue = 'sex')
plt.ylable('tip ($)')
plt.show()
```

```python
#Swarm plot
sns.swarmplot(x='day', y='tip', data=tips, hue = 'time')
plt.ylable('tip ($)')
plt.show()
```

```python
#Changing orientation 
sns.swarmplot(x='tip', y='day', data=tips, hue = 'sex', orient='h')
plt.ylable('tip ($)')
plt.show()
```

```python
plt.hist(tips['total_bill'], facecolor = 'green', edgecolor = 'white', bins = 12)
```

```python
plt.figure(figsize = (16,4))

plt.subplot(1,2,1)
sns.boxplot(x='day', y='tip', data=tips)
plt.ylabel('tip ($)')
plt.title('Boxplot of tips')

plt.subplot(1,2,2)
sns.vivolinplot(x='day', y='tip', data=tips)
plt.ylabel('tip ($)')
plt.title('Vivolin plot of tips')

plt.show()
```

```python
#Combining plots 

sns.vivolinplot(x='day', y='tip', data=tips, inner="box", color='lightgray')
sns.stripplot(x='day', y='tip', data=tips, size=4, jitter=True)
plt.ylabel('tip ($)')
plt.show()
```
